# GEOJSON dan Plotting dengan Geopandas

```python
neighborhoods_path = './geo.json'

# Read in the neighborhoods geojson file
neighborhoods = gpd.read_file(neighborhoods_path)

# Print the first few rows of neighborhoods
print(neighborhoods.head())

# Plot the neighborhoods, color according to name and use the Dark2 colormap
neighborhoods.plot(column = 'neighborhood', cmap = 'Dark2')

# Show the plot.
plt.show()
```

```python
# Print the first row of neighborhoods GeoDataFrame and the crs
print(neighborhoods.head(1))
print(neighborhoods.crs)

# Convert the crs to epsg:3857
neighborhoods.geometry = neighborhoods.geometry.to_crs(epsg = 3857)
                        
# Print the first row of neighborhoods GeoDataFrame and the crs again
print(neighborhoods.head(1))
print(neighborhoods.crs)
```

```python
bronx_polygon = bronx_polygon.to_crs(epsg = 4326)
bronx_polygon.crs
```

```python
bronx_centroid = bronx_polygon.geometry.centroid
bronx_polygon['center'] = bronx_centroid
bronx_polygon
```

```python
# Create allerton_center from the urban_polygon center
allerton_center = bronx_polygon.center[0]

# Print allerton_center
print(allerton_center)

# Create array for folium called allerton_location
allerton_location = [allerton_center.y, allerton_center.x]

# Print urban_location
print(allerton_location)
```

```python
# Construct a folium map with allerton_location
downtown_map = folium.Map(location = allerton_location, zoom_start = 12)

# Display the map
display(downtown_map)
```

```python
# Draw our neighborhood: Bronx
folium.GeoJson(bronx_polygon.geometry).add_to(downtown_map)

# Display the map
display(downtown_map)
```

```python
# Iterate through the art_dist_meters and print each part of tuple returned
for row in art_dist_meters.iterrows():
    print('first part: ', row[0])
    print('second part: ', row[1])
```

```python
art_dist_meters = art_dist_meters.to_crs(epsg = 4326)
```

```python
# Create a location and marker with each iteration for the downtown_map
for row in art_dist_meters.iterrows():
    row_values = row[1].geometry
    location = [row_values.y, row_values.x]
    marker = folium.Marker(location = location)
    marker.add_to(downtown_map)

# Display the map
display(downtown_map)
```

```python
# Create popups inside the loop you built to create the markers
for row in art_dist_meters.iterrows():
    row_values = row[1].geometry
    location = [row_values.y, row_values.x]
    popup = (str(row[1].NAME) + " : " + str(row[1].neighborhood_left))
    marker = folium.Marker(location = location, popup = popup)
    marker.add_to(downtown_map)

# Display the map.
display(downtown_map)
```
